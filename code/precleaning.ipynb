```python
import pandas as pd
```

# 동대문 전처리

```python
kakao = pd.read_csv(r'C:\Users\aug19pm7\Desktop\데이터분석\final\kakao_dongdaemun.csv')
kakao
```

```python
# 칼럼 없애기
kakao = kakao.drop(['Unnamed: 0'], axis = 1)
kakao
```

```python
# 'Store' 칼럼에서 공백 제거
kakao['Store'] = kakao['Store'].str.replace(" ", "")
kakao
```

```python
# 'Review_text' 칼럼에서 '정보 없음'이 아닌 행만 선택
kakao = kakao[kakao['Review_text'] != '정보 없음']
kakao
```

```python
# 'Store' 칼럼에서 '파리바게뜨' 또는 '뚜레쥬르'를 제외한 것만 남기기
kakao = kakao[~(kakao['Store'].str.contains('파리바게뜨') | kakao['Store'].str.contains('뚜레쥬르'))]
kakao
```

```python
# 'Time'칼럼 추가해서 'None' 으로 처리하기
kakao['Time'] = None
kakao
```

```python
# 저장
kakao.to_csv('kakao_dongdaemun.csv')
```
